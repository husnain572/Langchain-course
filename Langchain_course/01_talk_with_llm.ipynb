```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai_api_key=os.environ['OPENAI_API_KEY']
```

## Completion Model

```python
from langchain_openai import OpenAI
llm=OpenAI()
response=llm.invoke("Tell me about Imran Khan")
```

```python
print(response)
```

```python
for chunk in llm.stream("Tell me about Imran Khan"):
    print(chunk, end="",flush=True)
```

```python
creative_llm=OpenAI(temperature=0.9)
response=creative_llm.invoke("Tell me about Imran Khan")
print(response)
```

## Chat Model

```python
from langchain_openai import ChatOpenAI
chatModel=ChatOpenAI(model="gpt-3.5-turbo-0125")
messages=[('system','You are a historian expert in the Kennedy family'),
          ('human','Tell me about JFK.')]
response=chatModel.invoke(messages)
print(response.content)
print(response.response_metadata)
print(response.schema())
```

## Alternative LLM

```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai_api_key=os.environ['GROQ_API_KEY']
```

```python
from langchain_groq import ChatGroq
llamaChatModel=ChatGroq(
    model="llama-3.1-8b-instant"
)
messages=[('system','You are a historian expert in the Kennedy family'),
          ('human','Tell me about JFK.')]
response=llamaChatModel.invoke(messages)
print(response.content)
```
